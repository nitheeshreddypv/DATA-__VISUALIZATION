```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('iris.csv')
df.head()
```

```python
#pie chart of the classes
class_counts=df['variety'].value_counts()
labels=class_counts.index
values=class_counts.values
plt.figure(figsize=(10/2,10/2))
plt.pie(values,labels=labels,autopct='%1.0f%%')
plt.title('Distribution of classes pie chart')
plt.show()
```

```python
# Bar graph
plt.figure(figsize=(8, 6))
class_counts.plot(kind='bar', color='skyblue', edgecolor='k')
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Bar Graph of Class Distribution')
plt.xticks(rotation=0)
plt.show()
```

```python
#line plot
df2=pd.read_csv('dailymintemperatures.csv')
df2.tail(5)
plt.plot(df2['Date'][:7],df2['Temp'][:7],marker='o',color='g')
plt.xlabel('Date')
plt.ylabel('Temp in celsius')
plt.title('line plot of the temperatures')
plt.show()
```

```python
#boxplot of temperatures
plt.boxplot(df2['Temp'],vert=False)
plt.title('box plot of the temperatures ')
plt.show()
```

```python
# histogram of the sepallength vs the classes
for x in ('sepal.length','sepal.width'):
    plt.figure(figsize=(7,8))
    plt.hist(df['sepal.length'],bins=5,color='y',alpha=1)
    plt.x_label='sepallength'
    plt.y_label='frequency'
    plt.title('histogram of the :'+x)
    plt.show()
```

```python
# scatter plot b/w the sepal length and sepal width
plt.figure(figsize=(8,8))
plt.scatter(df['sepal.length'],df['sepal.width'],marker='o',color='g')
plt.xlabel='sepallength'
plt.ylabel='sepalwidth'
plt.title='scatterplot of the iris dataset'
plt.show
```
